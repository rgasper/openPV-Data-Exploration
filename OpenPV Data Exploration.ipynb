```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
openPV = pd.read_csv('pv_install_zip.csv')
```

```python
openPV['date_installed'] = pd.to_datetime(openPV['date_installed'], errors='coerce',infer_datetime_format=True)
```

```python
years = openPV['date_installed'].dt.year.value_counts().index.astype(int)
```

```python
years_dat = pd.DataFrame(index=years, columns=['avg_price'])
years_dat = years_dat.sort_index()
```

```python
temp_year = openPV.loc[openPV['date_installed'].dt.year == 2002]
temp_year.loc[temp_year['install_type'].str.lower() == 'residential']
```

```python
for year in years:
    temp_year = openPV.loc[openPV['date_installed'].dt.year == year]
    avg_p = temp_year['cost_per_watt'].mean()
    std_p = temp_year['cost_per_watt'].std()
    avg_s = temp_year['size_kw'].mean()
    tot_s = temp_year['size_kw'].sum()
    res_p = temp_year.loc[temp_year['install_type'].str.lower() == 'residential']['cost_per_watt'].mean()
    com_p = temp_year.loc[temp_year['install_type'].str.lower() == 'commercial']['cost_per_watt'].mean()
    gov_p = temp_year.loc[temp_year['install_type'].str.lower() == 'government']['cost_per_watt'].mean()
    np_p = temp_year.loc[temp_year['install_type'].str.lower() == 'nonprofit']['cost_per_watt'].mean()
    res_s = temp_year.loc[temp_year['install_type'].str.lower() == 'residential']['size_kw'].sum()
    com_s = temp_year.loc[temp_year['install_type'].str.lower() == 'commercial']['size_kw'].sum()
    gov_s = temp_year.loc[temp_year['install_type'].str.lower() == 'government']['size_kw'].sum()
    np_s = temp_year.loc[temp_year['install_type'].str.lower() == 'nonprofit']['size_kw'].sum()
    
    years_dat.at[year,'avg_price'] = avg_p
    years_dat.at[year,'std_price'] = std_p
    years_dat.at[year,'avg_size'] = avg_s
    years_dat.at[year,'tot_size'] = tot_s
    years_dat.at[year,'res_price'] = res_p; years_dat.at[year,'res_size'] = res_s
    years_dat.at[year,'com_price'] = com_p; years_dat.at[year,'com_size'] = com_s
    years_dat.at[year,'gov_price'] = gov_p; years_dat.at[year,'gov_size'] = gov_s
    years_dat.at[year,'np_price'] = np_p; years_dat.at[year,'np_size'] = np_s
```

```python
years_dat = years_dat.drop([1909,1969])
```

```python
X=years_dat.index; Y1=years_dat['avg_price']; Y2=years_dat['std_price'];
plt.errorbar(X,Y1,Y2,label='Avg $/W')
plt.ylabel('PV prices ($/W)')
plt.xlim[1995,2020]
plt.legend()
```

```python
X=years_dat.index;
fig, ax1 = plt.subplots()

color = 'black'
ax1.set_ylabel('Avg. Price ($/W)', color=color)
ax1.errorbar(X, years_dat['avg_price'], years_dat['std_price'],color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_facecolor('white')
ax1.set_xlim([1995,2015])
ax1.tick_params(axis='both',labelsize=12)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'blue'
ax2.set_ylabel('Total Installed Capacity (kW)', color=color)  # we already handled the x-label with ax1
ax2.semilogy(X,years_dat['tot_size'],'bo')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_xlim([1992,2015])
ax2.tick_params(axis='both',labelsize=12)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
years_res = pd.DataFrame(index=years, columns=['avg_price'])
years_res = years_res.sort_index()
years_com = pd.DataFrame(index=years, columns=['avg_price'])
years_com = years_com.sort_index()
years_gov_np = pd.DataFrame(index=years, columns=['avg_price'])
years_gov_np = years_gov_np.sort_index()

years_res.name = 'Res PV years'
years_com.name = 'Com PV years'
years_gov_np.name = 'Gov/NP PV years'
```

```python
dfs = [[resPV,years_res], [comPV,years_com], [gov_np_PV,years_gov_np]]
for df in dfs:
    for year in years:
        temp_year = df[0].loc[df[0]['date_installed'].dt.year == year]
        avg_p = temp_year['cost_per_watt'].mean()
        std_p = temp_year['cost_per_watt'].std()
        tot_s = temp_year['size_kw'].sum()
        avg_sol = temp_year['annual_insolation'].mean()
        std_sol = temp_year['annual_insolation'].std()
        
        df[1].at[year,'avg_price'] = avg_p
        df[1].at[year,'std_price'] = std_p
        df[1].at[year,'tot_size'] = tot_s
        df[1].at[year,'avg_sol'] = avg_sol
        df[1].at[year,'std_sol'] = std_sol
```

```python
X=years_res.index;
fig, ax1 = plt.subplots()

color = 'black'
fig.suptitle('Price and Total Capacity of USA PV Installations by Type', fontsize = 14)
ax1.set_ylabel('Avg. Price (---) ($/W)', color=color,fontsize = 14)
ax1.errorbar(X, years_res['avg_price'], years_res['std_price'],color='r',label='Residential',lw=5)
ax1.errorbar(X, years_com['avg_price'], years_com['std_price'],color='b',label='Commercial',lw=3)
ax1.errorbar(X, years_gov_np['avg_price'], years_gov_np['std_price'],color='g',label='Gov./ NP',lw=2)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_facecolor('white')
ax1.set_xlim([1992,2015])
ax1.set_ylim([2,20])
ax1.set_facecolor([0.25,0.25,0.25])
ax1.tick_params(axis='both',labelsize=12)
ax1.legend(fontsize=14, loc = 'upper left')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('Total Capacity (o) (kW)', color=color, fontsize = 14)  # we already handled the x-label with ax1
ax2.semilogy(X,years_res['tot_size'],'ro')
ax2.semilogy(X,years_com['tot_size'],'bo')
ax2.semilogy(X,years_gov_np['tot_size'],'go')
ax2.tick_params(axis='y', labelcolor=color)
ax2.tick_params(axis='both',labelsize=12)
ax2.set_xlim([1992,2015])
ax2.set_ylim([1e0,1e7])
ax2.tick_params(axis='both',labelsize=12)

#fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
fig.savefig('price_errplot.png', bbox_inches='tight')
```
